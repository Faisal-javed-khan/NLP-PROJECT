```python
import spacy
```

```python
nlp=spacy.load("en_core_web_sm")
```

```python
doc=nlp(u"hello i am engineer ")


# #"""u before a string (such as u"hi i am islamian")
# indicates that the string is a Unicode string literal.
#  In Python 2, there are two types of strings: ASCII strings
#  (plain strings) and Unicode strings. ASCII strings can only represent
#  characters from the ASCII character set (which includes basic Latin
#  characters and some common symbols)
# , while Unicode strings can represent characters from a much larger
#  set, including characters from various languages and special symbols."""
```

```python
doc.text
```

```python
doc[0]
```

```python
doc[0].pos_, doc[0].text
```

```python
spacy.explain("INTJ")
```

```python
for w in doc:
  print(w.text ,">>>>>", w.pos_ ,w.tag_, spacy.explain(w.tag_))
```

```python
doc[0].tag_
```

```python
spacy.explain("VB")
```

```python
doc1=nlp(u"i left the room")
doc2=nlp(u"on my left side")
```

```python
for w in doc1:
  print(w.text ,">>>>>", w.pos_ ,w.tag_, spacy.explain(w.tag_))
```

```python
for w in doc2:
  print(w.text ,">>>>>", w.pos_ ,w.tag_, spacy.explain(w.tag_))
```
